а)Визначаємо подій від противного, тобто визначаємо йомовірність не спрацювання першого та другого сигналізатора.(А1,А2)
Визначаємо подію що 2 сигналізатора не спрацюють одночасно.(А)
Подія коли спрацює хоча б 1 силналізатор дорівню 1-А

### Функція протилежної події

```python
def theorem_rev(*args):
    A = 1
    for i in range(len(args)):
        A *= args[i]
    return 1-A
```

```python
A1 = 1-0.95
A1
```

```python
A2 = 1-0.9
A2
```

```python
A = theorem_rev(A1,A2)
A
```
