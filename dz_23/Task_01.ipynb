Оскільки , сукупність подій незалежні та несумісні використовуємо теорему множення.

##  Функція теореми множення 

```python
def theorem_mult(*args):
    A = 1
    for i in range(len(args)):
        A *= args[i]
    return A
```

```python
A = theorem_mult(A1,A2,A3)
```
