### Пояснення розв'язку


Для того щоб оцінити ступінь ризу можно скористатись методом лінійної регресії , зокрема обчислити середньо квадратичне відхилення , яке  характеризує розсіяння значень навколо центру розподілу: більшому значенню стандартного відхилення відповідає більший їх розкид, тобто ступінь ризику. Для обчислення СКВ потрібно вичслити Математичне сподівання та Дисперсію , які розраховуються за формулами:

###### Математичне сподівання для груп товарів X та Y

```python
M_X = 1000*0.5+1500*0.3+2000*0.2
M_Y = 1000*0.4+1500*0.4+1750*0.2
```

###### Дисперсія для груп товарів X та Y

```python
D_X = (1000**2*0.5+1500**2*0.3+2000**2*0.2) - M_X**2
D_Y = (1000**2*0.4+1500**2*0.4+1750**2*0.2) - M_Y**2
```

######  СКВ

```python
from math import sqrt
```

```python
СКВ_X = sqrt(D_X)
СКВ_Y= sqrt(D_Y)
```

```python
СКВ_X
```

```python
СКВ_Y
```

Отже на основі даних розрахунків можно зробити висновок, що _СКВ_X > СКВ_Y_, отже і ступінь ризику буде вищий.

__Для продукції є рекомендований випуск товарів групи Y__
