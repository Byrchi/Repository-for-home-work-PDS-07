### Задача

####  Для виішення даної задачі спочатку визначемо основні  показники :
- Р - йомовірність гарантійного терміну
- q - йомовырнысть выдмови ( 1- Р)
- n - кількість телевізорів у вибірці
- межі вибірки [85; 90]

```python
P = 0.9
```

```python
q = 1 - P
```

```python
n = 100
```

Визначимо аналытичны показни( мат. очікування, дисперсія, СКВ)

```python
M_x = P * n
```

```python
D_x =  P * n * q
```

```python
σ_x = D_x**0.5
```

Використовуючи  центральну граничну  теорію розраховуємо:

```python
ε = 95 - M_x
ε
```

Використовуючі нерівність Чебишова , підставляємо всі значення в формулу


```python
P_x = 1 - D_x/ε**2
P_x
```
