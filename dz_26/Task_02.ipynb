##### Визначемо вибірку змінною х

```python
x = [10,13,10,9,9,12,12,6,7,9,8,9,11,9,14,13,9,8,8,7,10,10,11,11,11,12,8,7,9,10,14,13,8,8,9,10,11,11,12,12] 
len(x)
```

З данної вибірки отримуємо унікальні значення та підраховуємо їх кількість 

```python
unique_values = list(set(x))
value_counts = {value: x.count(value) for value in unique_values}

value_counts
```

На основі отриманих данних можемо побудувати дискретний статистичний розподіл та  полігон частот 
|X_i|6|7|8|9|10|11|12|13|14|
|:-|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|-:|
|N_i|1|3|6|8|6|6|5|3|2|

```python
import matplotlib.pyplot as plt
import numpy as np
x = value_counts.keys()
y = value_counts.values()
plt.title('Полігон частот')
plt.xlabel('Числа')
plt.ylabel('Частоти')
plt.plot(x,y, marker='o')
plt.show()
```

```python
total_count = len(x)
P_x = {value: count / total_count for value, count in value_counts.items()}
P_x
```

```python
import numpy as np
x = np.sort(x)
n = x.size
y = np.arange(1,n+1)/n

plt.step(x,y)
plt.xlabel('Числа')
plt.ylabel('F(x)')
plt.title('Емпірична функція розподілу')
plt.show()
```

```python
max_count = max(value_counts.values())
mode = []

for value, count in value_counts.items():
    if count == max_count:
        mode.append(value)

print("Мода:", mode)
```

```python
sorted_x = sorted(x)
n = len(sorted_x)
if n % 2 == 0:
    median = (sorted_x[n // 2 - 1] + sorted_x[n // 2]) / 2
else:
    median = sorted_x[n // 2]
print(f"Медіана: {median}")
```
