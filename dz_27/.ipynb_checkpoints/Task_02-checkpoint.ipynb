## Задача № 2 

####  В даній задачі також використовується закон розподілу Пуасона

Отже визначаємо змінні та підставляєм у формулу

```python
from math import sqrt
def analitic_indicators(n, q):
    M_x =round(n * q)
    D_x = round(M_x)
    S_x  = round(sqrt(D_x))
    return M_x, S_x
```

```python
n = 10000
```

```python
p = 0.9999
```

```python
q = 1 - p
round(q,4)
```

```python
answer2 = analitic_indicators(n, q)
answer2
```

```python
print(f"Математичне сподівання : {answer2[0]}","\n" f"Середнє квадратичне відхилення : {answer2[1]}")
```
