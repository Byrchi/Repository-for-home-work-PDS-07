#### Оскільки в даній задачі кількість випробуваннь дуже велика ( n ),  а ймовірність настання події ( р ) дуже мала - розподіл Пуасона

```python
n  = 10000
```

```python
p = 0.0007
```

Згідно з розподіла Пуасона:
-  M_x = n*p = D_x
-  S_X = sqrt(D_x)

```python
from math import sqrt
def analitic_indicators(x, y):
    M_x = round(x * y)
    D_x = round(M_x)
    S_x  = round(sqrt(D_x),4)
    return M_x, S_x
```

```python
answer = analitic_indicators(n, p)
```

### Відповідь

```python
print(f"Математичне сподівання : {answer[0]}","\n" f"Середнє квадратичне відхилення : {answer[1]}")
```
