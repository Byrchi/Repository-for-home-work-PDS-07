### Задача

#### По умові данної задачі нам не відома дисперсія отже будемо використовувати формулу для визначення інтервальної оціки :

Для вирішення даної задачі нам потрібно знайти 2 невідомі: 
- виправлене середнє квадратичне значення 
- x_mean - Середнє значення вибірки

_Визначаємо змінні_

```python
X = {1:2, 2:3 , 4:10, 6:3, 7:2}
```

```python
N = sum(X.values())
```

```python
x_mean = sum(k * v for k, v in X.items()) / N
 
print("Середнє значення вибірки : " ,x_mean)
```

Обраховуємо виправлення середнє квадратичне  значення 

```python
from math import sqrt
S = sqrt(sum(v * (k - x_mean) ** 2 for k, v in X.items()) / (N - 1))
S
```

Вираховуваємо точність оцінки де t = 2.093, оскільки беремо показник з таблиці при надійності у=0,95, та об'єму вибірки = 20

```python
t = 2.093
```

```python
δ = t*S/sqrt(N)
δ
```

```python
θ1 =round(x_mean - δ,3) 
```

```python
θ2 = round(x_mean + δ,3) 
```

### Відповідь 

```python
print(f"Інтервальна оцінка для СКВ {θ1} > θ > {θ2}")
```
