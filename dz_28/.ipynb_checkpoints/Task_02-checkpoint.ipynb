```python
t = 1.96
```

```python
p1 = 1/(t**2+n)*(m + t**2/2 - t*sqrt((m*(n-m)/n)+t**2/4))

p2 = 1/(t**2+n)*(m + t**2/2 + t*sqrt((m*(n-m)/n)+t**2/4))
p1,p2
```

```python
print(f"Інтервальна оцінка для ймовірності настання події А {round(p1,3)} < p < {round(p2,3)}")
```
