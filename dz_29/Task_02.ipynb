Оскільки Х розподіленна рівномірно то  щільність розподілу 

Якщо випадкова величина X розподілена рівномірно, то її щільність розподілу можна знайти за формулою f(x) = 1/(b-a), де a та b - це межі інтервалу, на якому розподілена випадкова величина

З умови задачі ми знаємо, що P(X>=3) = 0.4. Оскільки X розподілена рівномірно, то ймовірність того, що X приймає значення більше або рівне 3 дорівнює відношенню довжини відрізка [3, b] до довжини відрізка [a, b]

(b-3)/(b-a)=0.4

Mатематичне сподівання рівномірно розподіленої випадкової величини дорівнює середині інтервалу, на якому вона розподілена

(a+b)/2 = 2

Розв'яжемо систему рівнянь


(b-3)/(b-a)=0.4

(a+b)/2 = 2

(b-3) = 0.4(b-a)

a+b = 4

0.6b = 3 - 0.4a

a = 4-b

0.6b = 3 - 0.4(4-b)

0.2b = 1.4

b = 7

a = 4-7

a = - 3

Отже, значення a = -3 та b = 7.

Тепер можемо обчислити щільність розподілу f(x):

f(x) = 1 / (b - a), для a ≤ x ≤ b

f(x) = 1 / (7 - (-3)), для -3 ≤ x ≤ 7

f(x) = 1 / 10, для -3 ≤ x ≤ 7


Отже, межі розподілу X: a = -3 та b = 7. Щільність розподілу X дорівнює f(x) = 1/(b - a) для x ∈ [a, b] та f(x) = 0 для x ∉ [a, b]. Підставляючи значення a та b, ми отримуємо

__f(x) = 0.1 для x ∈ [-3, 7]__

__f(x) = 0 для x ∉ [-3, 7].__
