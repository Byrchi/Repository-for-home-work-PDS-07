Висуваємо гіпотезу __Н0 = що даний розподіл явлюється нормальним.__

Для того щоб підтвердити або спростувати данну гіпотезу потрібно розрахувати Кемп за формулою, та співставити з Ккр

```python
a_i = [[0,24],[24,48],[48,72],[72,96],[96,120],[120,144],[144,168],[168,192],[192,216]] # ВИзначаємо інтервали а_і
```

```python
n_i = [1, 2, 4, 6, 12, 16, 6, 2, 1]
```

```python
x_i = [] 
```

```python
for i in a_i:
    val = int(sum(i)/len(i))
    x_i.append(val)
```

```python
x_i  # Розраховуємо середине значення інтервалу
```

```python
n = 50
```

```python
sum_val = 0
for i in range(len(x_i)):
    sum_val += x_i[i]*n_i[i] 
x_av = sum_val/n
```

```python
x_av #  Розрахували середнэ значення х_штрих
```

```python
sum_val_D = []
for i in range(len(x_i)):
    el = n_i[i]*(x_i[i]- x_av)**2
    sum_val_D.append(el)
D = sum(sum_val_D)/n
```

```python
D # Розраховуємо дисперсію
```

```python
from  math import sqrt
S_av = sqrt(D)
S_av # Середнє квадратичне відхилення 
```

```python
z_i = [] # створюємо новий список для інтервалів (z_i-1  -  z_i)
for item in (a_i):
    z_i1_i = []
    for i in item:
        z_i1_i.append(round((i-x_av)/S_av, 2))
    z_i.append(z_i1_i) # В кожному вкладеному списку розраховуємо по формулі інтервали z_i 
z_i[0][0] = '-∞'
z_i[-1][-1] = '∞' # перший елемент в першому інтервалі та останній елемент в останньому інтервалі міняємо на -∞ та ∞ відповідно.
```

```python
z_i
```

```python
Φ_z_i = [[-0.5, -0.4913],   # Кожне занчення з попереднього списку переписую з таблиці Лапласа
          [-0.4913,-0.4599],
          [-0.4599,-0.3686 ],
          [-0.3686, -0.1879], 
          [-0.1879, 0.0557], 
          [0.0557, 0.2794], 
          [0.2794, 0.4192], 
          [0.4192, 0.4793], 
          [0.4793, 0.5], ]
```

```python
n_av = [] # Розраховую n-штрих  по кожному інтервалі
for i in Φ_z_i:
    n_av.append(round((i[1]-i[0])*n,3))
n_av
```

```python
K = []   # Розраховую Кемп по формулі, беручи суму значень ( Ni-N_середнє)**2/Ni 
for i in range(len(n_i)):
    K.append(((n_i[i]-n_av[i])**2)/n_i[i])
Kemp = round(sum(K),2)
print("Kemp = ", Kemp)
```

```python
Kkr = 16.8 # З талиці виписую Ккр при  значущості 0,01 та ступені вільності 6 
```

```python
# Порівнюю К емпіричне та К критичне , якщо К емпіричне менше за К критичне , тоді гіпотеза приймається , якщо ні то відхиляється .
if Kemp > Kkr:
    print(' Гіпотеза Н0 приймається')
else:
    print(' Гіпотеза Н0 відхиляється') 
```
