```python
a_i = [[0,24],[24,48],[48,72],[72,96],[96,120],[120,144],[144,168],[168,192],[192,216]] #   _
```

```python
n_i = [1, 2, 4, 6, 12, 16, 6, 2, 1]
```

```python
x_i = [] 
```

```python
for i in a_i:
    val = int(sum(i)/len(i))
    x_i.append(val)
```

```python
x_i  #    
```

```python
n = 50
```

```python
sum_val = 0
for i in range(len(x_i)):
    sum_val += x_i[i]*n_i[i] 
x_av = sum_val/n
```

```python
x_av #     _
```

```python
sum_val_D = []
for i in range(len(x_i)):
    el = n_i[i]*(x_i[i]- x_av)**2
    sum_val_D.append(el)
D = sum(sum_val_D)/n
```

```python
D #  
```

```python
from  math import sqrt
S_av = sqrt(D)
S_av #    
```

```python
z_i = [] #      (z_i-1  -  z_i)
for item in (a_i):
    z_i1_i = []
    for i in item:
        z_i1_i.append(round((i-x_av)/S_av, 2))
    z_i.append(z_i1_i) #         z_i 
z_i[0][0] = '-'
z_i[-1][-1] = '' #              -   .
```

```python
z_i
```
