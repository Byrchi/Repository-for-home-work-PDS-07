```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

___

# Побудова датасету

```python
cars = pd.read_excel('info_cars.xlsx', sheet_name='auto')
cars
```

___

# Побудова лінійного графіку залежності Вартість/Споживання палива

```python
x_price = list(cars['вартість'])
y_fuel_consumption = list(cars['споживання_палива'])
```

```python
plt.plot(x_price, y_fuel_consumption, linewidth=1 , markersize=3, color="brown" )
plt.xlabel('Вартість')
plt.ylabel('Споживання_палива')
plt.title('Залежність Вартість/Споживання палива')
plt.grid(True)
```

___


# Побудова точкового графіку залежності Вартість/Максимальна_швидкість

```python
z_speed = cars['максимальна_швидкість']
```

```python
plt.scatter(x_price, y_fuel_consumption ,  linewidth=1, marker='o')
plt.xlabel('Вартість')
plt.ylabel('максимальна_швидкість')
plt.title('Залежність Вартість/Максимальна_швидкість')
plt.grid(True)
```

___

# Побудова точкового графіку залежності Вартість/Вантажопідйомність

```python
k_load_capacity = cars['вантажопідйомність']
```

```python
plt.scatter(x_price, k_load_capacity ,  linewidth=1,  marker='o', s=50, alpha=0.8, edgecolors='black', linewidths=1)
plt.xlabel('Вартість')
plt.ylabel('Вантажопідйомність')
plt.title('Залежність Вартість/Вантажопідйомність')
plt.grid(True)
```

___

# Побудова гістограми поділу автомобілів за швидкістю

```python
plt.hist(list(cars['максимальна_швидкість']),  bins=5, color='green', edgecolor='black')
plt.title('Гістограма')
plt.xlabel('Швидкість')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```
