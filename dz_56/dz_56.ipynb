```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

# Завантажуємо  зображення

```python
img = cv2.imread('cat.jpg')
plt.axis('off')
plt.imshow(img)
plt.show
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
plt.axis('off')
plt.show
```

```python
img_RGB.shape
```

# Декомпозиція RGB

```python
R = img_RGB.copy()
G = img_RGB.copy()
B = img_RGB.copy()
```

```python
R[:,:,(1,2)]=0
plt.imshow(R)
plt.show()
```

```python
G[:,:,(0,2)]=0
plt.imshow(G)
plt.show()
```

```python
B[:,:,(0,1)]=0
plt.imshow(B)
plt.show()
```

# Декомпозиція HSV

```python
img_HSV = cv2.cvtColor(img_RGB,cv2.COLOR_RGB2HSV)
H = img_HSV.copy()
S = img_HSV.copy()
V = img_HSV.copy()
```

```python
H[ : , : , (1,2)] = 150
H_RGB = cv2.cvtColor(H, cv2.COLOR_HSV2RGB)
plt.imshow(H_RGB)
plt.show
```

```python
S[ : , : , 0] = 179
S[ : , : , 2] = 255
S_RGB = cv2.cvtColor(S, cv2.COLOR_HSV2RGB)
plt.imshow(S_RGB)
plt.show
```

```python
V[ : , : , 0] = 0
V[ : , : , 1] = 0
V_RGB = cv2.cvtColor(V, cv2.COLOR_HSV2RGB)
plt.imshow(V_RGB)
plt.show
```

# Декомпозиція LAB

```python
img_LAB = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2LAB)
```

```python
L = img_LAB.copy()
a = img_LAB.copy()
b = img_LAB.copy()
```

```python
L[ : , : , (1,2)] = 60
L_RGB = cv2.cvtColor(L, cv2.COLOR_LAB2RGB)
L_GRAY = cv2.cvtColor(L_RGB, cv2.COLOR_RGB2GRAY)
plt.imshow(L_RGB,cmap='gray')
plt.show
```

```python
a[ : , : , 0] = 127
a[ : , : , 2] = 0
a_RGB = cv2.cvtColor(a, cv2.COLOR_LAB2RGB)
a_GRAY = cv2.cvtColor(a_RGB, cv2.COLOR_RGB2GRAY)
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['red', 'gray', 'green'], N=255)
plt.imshow(a_GRAY, cmap=cmap_a)
plt.show
```

```python
b[ : , : , 1] = 127
b[ : , : , 2] = 0
b_RGB = cv2.cvtColor(b, cv2.COLOR_LAB2RGB)
b_GRAY = cv2.cvtColor(b_RGB, cv2.COLOR_RGB2GRAY)
cmap_b = clr.LinearSegmentedColormap.from_list('custom blue', ['yellow', 'gray', 'blue'], N=255)
plt.imshow(b_GRAY, cmap=cmap_b)
plt.show
```
