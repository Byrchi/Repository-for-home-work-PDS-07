```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title = 'Image', image=None, size = 15):
    w, h = image.shape[0], image.shape[1]
    aspect_ration = w/h 
    plt.figure(figsize=(size*aspect_ration, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show
```

## Оригінальне зображення

```python
img =  cv2.imread('stop.jpg')
imshow('dsfs',img)
```

## Градація сірого

```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imshow('Градація сірого', grey)
```

## Фільтр Sobel

```python
kernel_X = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
kernel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
adges_x = cv2.filter2D(grey,-1, kernel_X)
adges_y = cv2.filter2D(grey,-1, kernel_y)
sobel = adges_x + adges_y
imshow('sobel', sobel)
```

## Фільтр Canny

```python
canny = cv2.Canny(grey, 90 , 200 )
imshow('Canny',canny)
```

## Фільтр Roberts

```python
kernel_X = np.array([[-1,0,],[0,1]])
kernel_y = np.array([[0,-1,],[1,0]])
adges_x = cv2.filter2D(canny,-1, kernel_X)
adges_y = cv2.filter2D(canny,-1, kernel_y)
roberts = adges_x + adges_y
imshow('roberts', roberts)
```

## Висновок

Почергове застосування фільтрів Собеля, Кенні та Робертса до зображення може призвести до підкреслення різних аспектів та особливостей.

1.Фільтр Собеля акцентує границі та градієнти, визначаючи вертикальні та горизонтальні зміни.

2.Фільтр Кенні виявляє та точно визначає границі, включаючи етап відбору найсильніших границь та приглажування слабших.

3.Фільтр Робертса використовує прості матриці для виявлення границь, але може бути менш ефективним порівняно з іншими фільтрами.
